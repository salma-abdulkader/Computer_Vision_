```python
import cv2
from ultralytics import YOLO
```

```python
model=YOLO(r"C:\Users\DELL\Downloads\best.pt")
```

```python
# video_path=r"C:\Users\DELL\Downloads\WhatsApp Video 2024-10-30 at 7.36.12 PM.mp4"
video_path=0
```

```python
cap=cv2.VideoCapture(video_path)
names=model.model.names
```

```python
while cap.isOpened():
    success,frame=cap.read()
    frame=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    if success:
        results=model(frame)
        for result in results[0].boxes.cpu().numpy():
            x1,y1,x2,y2=map(int,result.xyxy[0])
            cls=names[int(result.cls[0])]
            conf=result.conf[0].round(2)
            cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),2) 
            cv2.putText(frame,cls,(x1,y1),fontFace=cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,fontScale=1,color=(0,255,0))
            cv2.putText(frame,str(conf),(x2,y2),fontFace=cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,fontScale=1,color=(0,255,0))
        frame=cv2.cvtColor(frame,cv2.COLOR_RGB2BGR)
        cv2.imshow("Yolo v11s",frame)
        if cv2.waitKey(1) & 0xFF==ord("q"):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows            
```
